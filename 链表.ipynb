## 单向链表

```python
class Node(object):
    """节点"""
    def __init__(self,elem):
        self.elem = elem
        self.next = None
    
class SingleLinkList(object):
    """单链表"""
    def __init__(self, node=None):
        self.__head = node     # __表示内部函数——私有属性head    
        
    def is_empty(self):
        """链表是否为空 """
        return self.__head == None
    
    def length(self):
        """链表长度""" 
        # Cur 游标，用来移动遍历节点
        Cur = self.__head
        # count记录数量
        count = 0
        while Cur !=None:
            count += 1
            Cur = Cur.next
        return count 
    
    def travel(self):
        """ 遍历整个链表"""
        # Cur 游标，用来移动遍历节点
        Cur = self.__head
        while Cur !=None:
            print(Cur.elem, end=' ')   # python3换行‘_’,python2换行‘，’
            Cur = Cur.next
        return 
                
    
    def add(self,item):
        """链表头部添加元素，头插法"""
        node = Node(item)
        node.next = self.__head
        self.__head = node 
    
    def append(self,item):
        """链表尾部添加元素,尾插法"""
        node = Node(item)
        if self.is_empty():
            self.__head = node
        else:
            # Cur 游标，用来移动遍历节点,游标也可以叫做指针
            Cur = self.__head
            while Cur.next !=None:
                Cur = Cur.next
            Cur.next = node
        
    
    def insert(self,pos, item):
        """指定位置添加元素
        :param pos  从0开始
        """
        if pos <= 0:
            self.add(item)
        elif pos > (self.length()-1):
            self.append(item)
        else:
            pre = self.__head   # 让pre等于第0个节点的数据
            count = 0
            while count < (pos-1):
                count +=1
                pre = pre.next
            ### 当循环退出后，开始添加操作
            node = Node(item)
            node.next = pre.next
            pre.next = node

    def remove(self,item):
        """删除节点 """
        Cur = self.__head
        pre = None
        while Cur != None: 
            if Cur.elem == item:
                # 先判断此节点是不是头节点，是头节点：
                if Cur == self.__head:
                    self.__head = Cur.next
                else:
                    pre.next = Cur.next
                break
            else:
                pre = Cur
                Cur = Cur.next
          
    
    def search(self,item):
        """查找节点是否存在 """
        Cur = self.__head
        while Cur != None: 
            if Cur.elem == item:
                return True
            else:
                Cur = Cur.next
        return False   
```
